### 2. Cell-type-specific pseudobulk generation 


**Input: ./data/snRNA_donor**

**Output: ./data/pb_ct**

```python
import pseudobulk_ct

# 1. 출력 경로 설정 (기존 root_path_data 활용)
pb_ct_path = f"../../data/pb_ct"

# 2. 분석 옵션 설정
# cellclass 기준으로 나누며, 대상 세포 타입들을 리스트로 지정합니다.
ct_col = 'cellclass'
cell_types = [
    "gabaergic", "glutamatergic", "oligodendrocyte",
    "microglia", "astrocyte", "polydendrocyte", "endothelia"
]

pseudobulk_ct.main(
    donor_metadata_path=donor_metadata_path,
    donor_adata_path=donor_adata_path,
    pb_ct_path=pb_ct_path,
    ct_col=ct_col,
    cell_types=cell_types
)
```
